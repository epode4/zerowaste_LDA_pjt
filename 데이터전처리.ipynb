### 정규식 적용

```python
# 문장 분리 함수
def separate_sentences(text):
    sentences = re.split(r'(?<=[.!?]) +', text)
    return sentences

# 정규식
def regular_expression(text):
    # URL 제거
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    text_without_url = url_pattern.sub('', text)

    # 한글, 영어, 띄어쓰기만 남기기
    clean_pattern = re.compile('[^a-zA-Z가-힣 ]')
    result = clean_pattern.sub('', text_without_url)

    return result
```

```python
# 문장 분리 후 정규식 적용해서 한글, 영어만 남도록 정제 함수
def apply_sep_regex(text):
    sentences = separate_sentences(text)
    reg_sentences=[]
    for s in sentences:
        reg_sentences.append(regular_expression(s))
    return reg_sentences
```

```python
df_cafe_norm['Content'] = df_cafe_norm['Content'].apply(lambda x: apply_sep_regex(x))
```

```python
# 빈 값이 들어간 Content None값 처리
def eliminate_empty_lines(text):
    non_empty = [line for line in text if line.strip()]
    return non_empty if non_empty else None

df_cafe_norm['Content'] = df_cafe_norm['Content'].apply(eliminate_empty_lines)
df_cafe_norm = df_cafe_norm.dropna(subset=['Content']).reset_index(drop=True)
```

## 토큰화

**SOYNLP의 Noun Extractor(v.2) 활용**

soynlp는 품사 태깅, 단어 토큰화들을 제공함.
비지도 학습으로 단어 토큰화 진행 - 데이터에서 자주 등장하는 단어들 (신조어 포함) 단어로 분석

```python
from soynlp.noun import LRNounExtractor_v2

noun_extractor = LRNounExtractor_v2(verbose=True, extract_compound=True)
```

#### 1. 각 게시글별 명사(+게시글별 빈도수) 추출

df['Content']는 각 게시글을 문장단위로 나뉜 elements를 가진 list

sents = df_blog.loc[idx,'Content']

sents를 noun_extractor에 넣어서 각 게시글별 명사와 빈도수 및 score를 추출

```python
def extract_nouns(df,freq):
    nouns=[]
    for idx in df.index:
        sents = df.loc[idx,'Content']
        try:
            nouns_per_posting = noun_extractor.train_extract(
            sents,
            min_noun_frequency=freq)
            nouns.append(nouns_per_posting)
        except:
            pass
    return nouns
```

```python
# minfreq=3 설정 이유 - 서론,본론,결론에 적어도 한번씩 등장하는 단어라면 3+ 예상함.
nouns_minfreq_blog = extract_nouns(df_cafe_norm,3)
nouns_minfreq_blog
```

```python
for word in zerowaste_stopwords:
    stopwords.append(word)
print(len(stopwords))
```

```python
include = ['낮', '땅', '밭', '꽃', '돌', '멋','맛', '폼', '물', '볕', '빛', '봄', '숲', '새', '산', '숨', '싹', '옷', '잎', '차', '흙', '힘']
```

```python
# 불용어 제거 함수 수정
def eliminate_stopwords(nouns):
    filtered = []
    for n in nouns:
        n_list = n.split(',')
        filtered_per_post = []
        for x in n_list:
            pattern = re.compile('[^a-zA-Z|ㄱ-ㅣ가-힣]')
            result = pattern.sub('', x)
            x = x.replace("'", "")
            try:
                x = x.replace(" ", "")
            except:
                pass
            if (len(x) > 1 or x in include) and (x not in stopwords):
                filtered_per_post.append(x)
        # 빈 값이 아닌 경우에만 추가
        if filtered_per_post:
            filtered.append(', '.join(filtered_per_post).strip(', '))
    return filtered
```

```python
df_cafe_norm_copy1['Content'] = df_cafe_norm_copy1['Content'].apply(lambda x: eliminate_stopwords(x))
```
